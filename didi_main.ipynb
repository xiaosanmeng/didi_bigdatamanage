```python
#导入PIL，os，glob
from PIL import Image
import os,glob
import pandas as pd
import time,datetime
```

```python
#导入数据
file=r'C:\Users\ASUS\Desktop\bigdatamanage\20201114\data.csv'
data=pd.read_csv(file,header=None)
```

```python
data.info()
```

```python
data=data.rename(columns={0:'0',1:'startt',2:'endt',3:'on_lon',4:'on_lat',5:'off_lon',6:'off_lat',7:'value'})
```

```python
data.drop(['0','value'],axis=1,inplace=True)
```

```python
data.describe()
```

```python
data.loc[:, 'localminute1'] = data['startt'].apply(lambda x :time.localtime(x))
data.loc[:, 'time_s'] = data['localminute1'].apply(lambda x :time.strftime("%Y-%m-%d %H:%M:%S", x))
```

```python
data.drop(['localminute1'],axis=1,inplace=True)
```

```python
data.loc[:, 'localminute2'] = data['endt'].apply(lambda x :time.localtime(x))
data.loc[:, 'time_e'] = data['localminute2'].apply(lambda x :time.strftime("%Y-%m-%d %H:%M:%S", x))
data.drop(['localminute2'],axis=1,inplace=True)
```

```python
#提取时间特征
data.loc[:, 'time_s'] = pd.to_datetime(data['time_s'],format='%Y-%m-%dT%H:%M:%S')

#获取秒
data.loc[:, 'second_s'] = data['time_s'].apply(lambda x: x.second)
#获取分
data.loc[:, 'minute_s'] = data['time_s'].apply(lambda x: x.minute)
#获取时
data.loc[:, 'hour_s'] = data['time_s'].apply(lambda x: x.hour)
#获取日
data.loc[:, 'date_s'] = data['time_s'].apply(lambda x: x.day)

#提取时间特征
data.loc[:, 'time_e'] = pd.to_datetime(data['time_e'],format='%Y-%m-%dT%H:%M:%S')

#获取秒
data.loc[:, 'second_e'] = data['time_e'].apply(lambda x: x.second)
#获取分
data.loc[:, 'minute_e'] = data['time_e'].apply(lambda x: x.minute)
#获取时
data.loc[:, 'hour_e'] = data['time_e'].apply(lambda x: x.hour)
#获取日
data.loc[:, 'date_e'] = data['time_e'].apply(lambda x: x.day)
```

```python
#时间切片
data['deal_dates'] = pd.to_datetime(data['time_s'])
data['date_min_interval']=data['time_s'].dt.minute//15
data['date_hour']=data['time_s'].dt.hour
data['period']=data['date_hour']*6+data['date_min_interval']
data.head()
```
